## Read Clean and Validate data
### Weight variable

```python
nsfg['birthwgt_lb1'].describe()
```

```python
# The values of 98 and 99 represent as invalid
pounds = nsfg['birthwgt_lb1'].replace([98,99], np.nan)
ounces = nsfg['birthwgt_oz1'].replace([98,99], np.nan)

# Converting all to pounds metric 
birth_weight = pounds + ounces/16
birth_weight
```

```python
nsfg['birthwgt_lb1'].value_counts().sort_index(), nsfg['birthwgt_oz1'].value_counts().sort_index()
```

```python
plt.hist(birth_weight.dropna(), bins=30)
plt.xlabel('Birth wright (lb)')
plt.ylabel('Fraction of births')
plt.show()
```

### Full term babies and single births
#### Babies born preterm < 37 weeks
#### Babies born long term >= 37 weeks

```python
nsfg['nbrnaliv'].value_counts()
```

```python
nsfg['nbrnaliv'].replace(8, np.nan, inplace=True)
nsfg['nbrnaliv'].value_counts()
```

```python
full_term = nsfg['prglngth'] >= 37
# Select the weights of full-term babies
full_term_weight = birth_weight[full_term]
full_term_weight.mean()
```

```python
# Filter full-term babies
full_term = nsfg['prglngth'] >= 37

# Filter single births
single = nsfg['nbrnaliv'] == 1

# Compute birth weight for single full-term babies
single_full_term_weight = birth_weight[single & full_term]
print('Single full-term mean:', single_full_term_weight.mean())

# Compute birth weight for multiple full-term babies
mult_full_term_weight = birth_weight[~single & full_term]
print('Multiple full-term mean:', mult_full_term_weight.mean())
```

# PMF Probability Mass Function
## The probability that you get exactly x
Represents the unique values in the dataset and how often each one appears.


```python
from empiricaldist import Pmf
```

```python
pmf_educ = Pmf.from_seq(educ, normalize=True)
pmf_educ.head()
```

```python
pmf_educ.bar()
plt.xlabel('Years of education')
plt.ylabel('PMF')
plt.show()
```

```python
# Compute the PMF for year
pmf_year = Pmf.from_seq(gss['year'], normalize=False)
pmf_year.head()
```

```python
pmf_year.bar()
plt.xlabel('Year')
plt.ylabel('PMF')
plt.show()
```

```python
age = gss['age']
pmf_age = Pmf.from_seq(age)
pmf_age.head()
```

```python
pmf_age.bar()
plt.xlabel('Age')
plt.ylabel('PMF')
plt.show()
```

# CDF Cumulative Distribution Function
## The probability that you get a value less than or equal to x

```python
from empiricaldist import Cdf
```

```python
cdf_age = Cdf.from_seq(age)
cdf_age.head()
```

```python
cdf_age.plot()
plt.xlabel('Age')
plt.ylabel('CDF')
plt.show()
```

```python
q = cdf_age[51]
print(q)
```

```python
p = 0.25
q = cdf_age.inverse(p)
print(q)
```

```python
p = 0.75
q = cdf_age.inverse(p)
print(q)
```

```python
income = gss['realinc']
cdf_income = Cdf.from_seq(income)
# Calculate the 75th percentile 
percentile_75th = cdf_income.inverse(0.75)

# Calculate the 25th percentile
percentile_25th = cdf_income.inverse(0.25)

# Calculate the interquartile range
iqr = percentile_75th - percentile_25th

# Print the interquartile range
iqr
```

```python
cdf_income.plot()
plt.xlabel('Income (1986 USD)')
plt.ylabel('CDF')
plt.show()
```

### Modeling Distributions

#### Normal distribution

```python
sample = np.random.normal(size=1000)
Cdf.from_seq(sample).plot()
```

```python
from scipy.stats import norm
xs = np.linspace(-3,3)
ys = norm(0.1).cdf(xs)
plt.plot(xs, ys, color='gray')
Cdf.from_seq(sample).plot(label='cdf')
plt.legend()
plt.show()
```

> If this was real data we would conclude that the normal distribution is a good model for the data.

### The bell curve 
#### PDF (Probability Density Function)

```python
xs = np.linspace(-3,3)
ys = norm(0.1).pdf(xs)
plt.plot(xs, ys, color='gray')
Pmf.from_seq(sample).plot(label='Pmf')
plt.legend()
plt.show()
```

> The pdf does not work very well

### KDE Kernel Density Estimation

```python
import seaborn as sns
sns.kdeplot(sample)
```

```python
xs = np.linspace(-3,3)
ys = norm(0.1).pdf(xs)
plt.plot(xs, ys, color='gray')
sns.kdeplot(sample)
plt.show()
```

> It's a way of getting from a PMF, a probability mass function, to a PDF, a probability density function.

#### 1. CDF for exploration
#### 2. PMF if there are a small number of unique values
#### 3. KDE id there are a lot of values

### Income distribution

```python
# Extract realinc log
income = gss['realinc']
log_income = np.log10(income)

# Compute mean and standard deviation
mean = log_income.mean()
std = log_income.std()
print(mean, std)

# Make a norm object
from scipy.stats import norm
dist = norm(0,1)
```

```python
# Evaluate the model CDF
xs = np.linspace(2, 5.5)
ys = dist.cdf(xs)

# Plot the model CDF
plt.clf()
plt.plot(xs, ys, color='gray')

# Create and plot the Cdf of log_income
Cdf.from_seq(log_income).plot()
    
# Label the axes
plt.xlabel('log10 of realinc')
plt.ylabel('CDF')
plt.show()
```

```python
# Evaluate the normal PDF
xs = np.linspace(2, 5.5)
ys = dist.pdf(xs)

# Plot the model PDF
plt.clf()
plt.plot(xs, ys, color='gray')

# Plot the data KDE
sns.kdeplot(log_income)

# Label the axes
plt.xlabel('log10 of realinc')
plt.ylabel('PDF')
plt.show()
```

### Simple regression

```python
from scipy.stats import linregress
subset = brfss.dropna(subset=['WTKG3', 'HTM4'])
xs = subset['HTM4']
ys = subset['WTKG3']
res = linregress(xs, ys)
res
```

#### Regression Line

```python
fx = np.array([xs.min(), xs.max()])
fy = res.intercept + res.slope * fx
```

```python
height_jitter = height + np.random.normal(0,2, size=len(brfss))
weight_jitter = weight + np.random.normal(0,2, size=len(brfss))
plt.plot(height_jitter, weight_jitter, 'o', markersize=1,alpha=0.02)
plt.plot(fx,fy,'-')
plt.axis([140, 200, 0, 160]) #Zoom
plt.xlabel('Height in cm')
plt.ylabel('Weight in kg')
plt.show()
```

```python
subset = brfss.dropna(subset=['WTKG3', 'AGE'])
xs = subset['AGE']
ys = subset['WTKG3']
res = linregress(xs, ys)
res
```

```python
fx = np.array([xs.min(), xs.max()])
fy = res.intercept + res.slope * fx
```

```python
age = brfss['AGE'] + np.random.normal(0, 0.5, size=len(brfss))
weight = brfss['WTKG3'] + np.random.normal(0, 2, size=len(brfss))

plt.plot(age, weight, 'o', markersize=1, alpha=0.2)
plt.plot(fx,fy,'-')
plt.xlabel('Age in years')
plt.ylabel('Weight in kg')
plt.show()
```

```python
subset = brfss.dropna(subset=['INCOME2', '_VEGESU1'])
xs = subset['INCOME2']
ys = subset['_VEGESU1']
res = linregress(xs,ys)
print(res)
```

```python
fx = np.array([xs.min(), xs.max()])
fy = res.intercept + res.slope * fx
x_jitter = xs + np.random.normal(0, 0.15, len(xs))

plt.clf()
plt.plot(x_jitter, ys, 'o', alpha=0.2)
plt.plot(fx, fy, '-', alpha=0.7)
plt.xlabel('Income code')
plt.ylabel('Vegetable servings per day')
plt.ylim([0, 6])
plt.show()
```

## Multivariate Thinking

```python
from scipy.stats import linregress
import statsmodels.formula.api as smf

# Run regression with linregress
subset = brfss.dropna(subset=['INCOME2', '_VEGESU1'])
xs = subset['INCOME2']
ys = subset['_VEGESU1']
res = linregress(xs,ys)
print(res)

# Run regression with StatsModels
results = smf.ols('_VEGESU1 ~ INCOME2', data = brfss).fit()
print(results.params)
```

### Multiple Regression

```python
gss = pd.read_hdf('gss.hdf5', 'gss')
gss.head()
```

```python
results = smf.ols('realinc ~ educ', data=gss).fit()
results.params
```

#### Adding Age variable

```python
results = smf.ols('realinc ~ educ + age', data=gss).fit()
results.params
```

```python
mean_income_by_age = gss.groupby('age')['realinc'].mean()
plt.plot(mean_income_by_age, 'o', alpha=0.5)
plt.xlabel('Age (Years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
gss['age2'] = gss['age']**2
model = smf.ols('realinc ~ educ + age + age2', data=gss)
results = model.fit()
results.params
```

### Income and education

```python
mean_income_by_educ = gss.groupby('educ')['realinc'].mean()
plt.plot(mean_income_by_educ, 'o', alpha=0.5)
plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

> It looks like the relationship between income and education is non-linear.

```python
import statsmodels.formula.api as smf

gss['educ2'] = gss['educ']**2

# Run a regression model with educ, educ2, age, and age2
results = smf.ols('realinc ~ educ + educ2 + age + age2', data=gss).fit()
results.params
```

>The slope associated with educ2 is positive, so the model curves upward. 

### Visualizing regression results

```python
df = pd.DataFrame()
df['age'] = np.linspace(18,85)
df['age2'] = df['age']**2

df['educ'] = 12
df['educ2'] = df['educ']**2
```

```python
pred12 = results.predict(df)

plt.plot(mean_income_by_age, 'o', alpha=0.5)
plt.plot(df['age'], pred12, label='High School')
plt.xlabel('Age (Years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
# predict
df['educ'] = 14
df['educ2'] = df['educ']**2
pred14 = results.predict(df)
# plot
plt.plot(mean_income_by_age, 'o', alpha=0.5)
plt.plot(df['age'], pred14, label='Associate')
plt.xlabel('Age (Years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
# predict
df['educ'] = 16
df['educ2'] = df['educ']**2
pred16 = results.predict(df)
# plot
plt.plot(mean_income_by_age, 'o', alpha=0.5)
plt.plot(df['age'], pred16, label='Bachelor')
plt.xlabel('Age (Years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
plt.plot(mean_income_by_age, 'o', alpha=0.5)
plt.plot(df['age'], pred16, label='Bachelor')
plt.plot(df['age'], pred14, label='Associate')
plt.plot(df['age'], pred12, label='High School')
plt.legend()
plt.xlabel('Age (Years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

#### Making predictions

```python
# Run a regression model with educ, educ2, age, and age2
results = smf.ols('realinc ~ educ + educ2 + age + age2', data=gss).fit()

# Empty DataFrame
df = pd.DataFrame()
df['educ'] = np.linspace(0,20)
df['age'] = 30
df['educ2'] = df['educ']**2
df['age2'] = df['age']**2

# Generate and plot the predictions
pred = results.predict(df)
pred.head()
```

```python
# Plot mean income in each age group
plt.clf()
mean_income_by_educ = gss.groupby('educ')['realinc'].mean()
plt.plot(mean_income_by_educ, 'o', alpha=0.5)

# Plot the predictions
pred = results.predict(df)
plt.plot(df['educ'], pred, label='Age 30')
plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')
plt.legend()
plt.show()
```

> Looks like this model captures the relationship pretty well.
